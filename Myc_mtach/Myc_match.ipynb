```python
import pandas as pd

# --- Step 1: Load target sequence ---
with open("myc_3utr.txt", "r") as f:
    myc_3utr = f.read().strip()

# --- Step 2: Load sequences from CSV ---
df = pd.read_csv("/home/yg2895/media/protein/yg2895/ASO-lib-20250102/motif_analyse/haplo-utr3-oligo-pool.csv")
sequences = df['Sequence'].tolist()

# --- Step 3: Function to get the longest exact matching substring ---
def longest_common_substring(s1, s2):
    m, n = len(s1), len(s2)
    longest = 0
    end_index_s1 = 0
    dp = [[0]*(n+1) for _ in range(m+1)]

    for i in range(m):
        for j in range(n):
            if s1[i] == s2[j]:
                dp[i+1][j+1] = dp[i][j] + 1
                if dp[i+1][j+1] > longest:
                    longest = dp[i+1][j+1]
                    end_index_s1 = i + 1
    return s1[end_index_s1 - longest:end_index_s1]

# --- Step 4: Get top 10 matches based on substring length ---
results = []
for seq in sequences:
    motif = longest_common_substring(myc_3utr, seq)
    results.append((seq, motif, len(motif)))

# --- Step 5: Sort and select top matches ---
results.sort(key=lambda x: x[2], reverse=True)
top_matches = results[:20]

# --- Step 6: Print only matched motifs ---
print("Top 10 matched motifs (longest exact substring):\n")
for i, (seq, motif, length) in enumerate(top_matches, 1):
    print(f"{i}. {motif}")
```

```python
import pandas as pd
from collections import defaultdict

# --- Step 1: Load target sequence ---
with open("myc_3utr.txt", "r") as f:
    myc_3utr = f.read().strip()

# --- Step 2: Load oligo sequences ---
df = pd.read_csv("/home/yg2895/media/protein/yg2895/ASO-lib-20250102/motif_analyse/haplo-utr3-oligo-pool.csv")
sequences = df['Sequence'].tolist()

# --- Step 3: Generate all substrings ≥12 nt from myc_3utr ---
min_len = 12
motif_counts = defaultdict(int)

seen_motifs = set()
for length in range(min_len, len(myc_3utr) + 1):
    for start in range(len(myc_3utr) - length + 1):
        motif = myc_3utr[start:start + length]
        if motif in seen_motifs:
            continue
        seen_motifs.add(motif)

        # --- Step 4: Count how many oligos contain this motif ---
        count = sum(1 for seq in sequences if motif in seq)
        if count > 0:
            motif_counts[motif] = count

# --- Step 5: Sort motifs by how many oligos they appear in ---
sorted_motifs = sorted(motif_counts.items(), key=lambda x: x[1], reverse=True)

# --- Step 6: Display top motifs ---
print("Top recurring motifs (length ≥12 nt):\n")
for i, (motif, count) in enumerate(sorted_motifs[:20], 1):
    print(f"{i}. Occurrence: {count} | Motif: {motif}")
```
