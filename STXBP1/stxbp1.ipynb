```python
import matplotlib.pyplot as plt
import numpy as np

# Initialize lists for storing ratios and errors
ls_ratio = []
ls_err = []

# Loop through rows in steps of 60
for x in range(0, 480, 60):
    # Extract WT rows
    stx_wt = stx_oligo.iloc[[x, x+2, x+4, x+6], :]
    # Extract SCB rows
    stx_scb = stx_oligo.iloc[[x+1, x+3, x+5, x+7], :]
    # Extract MUT rows
    stx_mut = stx_oligo.iloc[x+8:x+60]

    # Compute sk_rd for WT
    sk_rd_wt = (stx_wt.iloc[:,11:14].mean(axis=1)/stx_wt.iloc[:,8:11].mean(axis=1)).median()
    
    # Process MUT rows in blocks of 4
    for i in range(0, 52, 4):
        # Extract 4-row blocks from MUT
        stx_mut1 = stx_mut.iloc[i:i+4]
        
        # Compute sk_rd for MUT
        sk_rd_mut = (stx_mut1.iloc[:,11:14].mean(axis=1)/stx_mut1.iloc[:,8:11].mean(axis=1))
        
        # Calculate error and ratio
        err = (sk_rd_mut / sk_rd_wt).std()
        ratio = sk_rd_mut.median() / sk_rd_wt

        # Append results to lists
        ls_err.append(err)
        ls_ratio.append(ratio)

# Generate x-axis values dynamically to match the number of ratios
xlab = list(range(1, len(ls_ratio) + 1))

# Plotting
plt.rcParams["figure.figsize"] = [30, 5]
plt.rcParams["figure.autolayout"] = True

# Create scatter plot with error bars
plt.scatter(xlab, ls_ratio, label='Ratio (MUT/WT)', color='black')
plt.errorbar(xlab, ls_ratio, ls_err, linestyle='None', marker='*', color='blue', label='Error')

# Add horizontal line at y=1 for baseline
plt.axhline(y=1, color='r', linestyle='dashed', label='Baseline (y=1)')

# Customize plot
plt.xticks(np.arange(min(xlab), max(xlab) + 1, 1.0))
plt.xlabel("Positions")
plt.ylabel("Median Ratio (MUT/WT)")
plt.title("Mutant/WT Ratios with Error Bars")
# plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Initialize dictionaries for ratios
ls_ratios = {f"ls_ratio{i}": [] for i in range(1, 5)}
ls_ratios_1 = {f"ls_ratio{i}_1": [] for i in range(1, 5)}

# Loop through the main blocks
for x in range(0, 480, 60):
    stx_wt = stx_oligo.iloc[[x, x+2, x+4, x+6], :]
    stx_mut = stx_oligo.iloc[x+8:x+60]

    # Compute WT ratio
    hek_rd_wt = (stx_wt.iloc[:, 5:8].mean(axis=1) / stx_wt.iloc[:, 2:5].mean(axis=1)).median()

    # Determine which dictionary to use based on the range
    if x in range(0, 420):
        ratio_dict = ls_ratios
    else:
        ratio_dict = ls_ratios_1

    # Process mutant rows in blocks of 4
    for i in range(0, 52, 4):
        for j in range(4):  # Process each sub-block
            stx_mut_sub = stx_mut.iloc[i+j:i+j+1]
            hek_rd_mut = (stx_mut_sub.iloc[:, 5:8].mean(axis=1) / stx_mut_sub.iloc[:, 2:5].mean(axis=1)).median()
            ratio_dict[f"ls_ratio{j+1}" if x in range(0, 420) else f"ls_ratio{j+1}_1"].append(hek_rd_mut / hek_rd_wt)

# Prepare x-axis values
xlab = list(range(1, 92))
xlab2 = list(range(85, 98))

# Plotting
plt.rcParams["figure.figsize"] = [30, 5]
plt.rcParams["figure.autolayout"] = True

fig, ax = plt.subplots()

# Plot data
for i in range(1, 5):
    ax.scatter(xlab, ls_ratios[f"ls_ratio{i}"], label=f"Ratio {i} (Range 0-420)", s=50)
    ax.scatter(xlab2, ls_ratios_1[f"ls_ratio{i}_1"], label=f"Ratio {i} (Range 420-480)", s=50)

# Customize plot
plt.xticks(np.arange(min(xlab), max(xlab2)+1, 1.0))
plt.axhline(y=1, color='r', linestyle='dashed', label="Baseline (y=1)")
plt.xlabel("Positions")
plt.ylabel("Ratios (MUT/WT)")
plt.title("Mutant/WT Ratios by Range")
# plt.legend()
plt.show()
```
